SGD(stochastic gradient descent) is a method to find the minimum value of target loss function when there may exist many local minimum extreme value. The opposite direction of gradient is fastest decreasing direction of the loss function.


The following program implements a linear regression example with SGD. Mainly 1) fetch and pre-process the train data 2) build and train the model, i.e, to find the proper weights. 3) predict the result with test data.
